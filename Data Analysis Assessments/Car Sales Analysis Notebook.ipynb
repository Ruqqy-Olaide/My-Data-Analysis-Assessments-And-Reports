## **Specific Analysis**

```python
#5. Manufacturer with the highest and lowest average sales volume
#a. Highest
data.groupby('Manufacturer')['Sales_in_thousands'].mean().sort_values(ascending=False)
```

```python
#OR
pv1 = data.pivot_table(index='Manufacturer', values='Sales_in_thousands', aggfunc='mean')
pv1.sort_values(by='Sales_in_thousands', ascending=False)
```

```python
#b. Smallest
data.groupby('Manufacturer')['Sales_in_thousands'].mean().sort_values(ascending=True)
```

```python
#distribution of car prices in the dataset
pivot1 = data.pivot_table(index='Manufacturer', values='Price_in_thousands', aggfunc='sum')
pivot1

pivot1.plot(kind='bar', figsize=(8, 5))
plt.xlabel('Manufacturer')
plt.ylabel('Sum of prices')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
#8. Correlation matrix of numerical variables in the dataset
data_correlation = data.corr(numeric_only=True)
data_correlation
```

```python
#Plotting the correlation
plt.figure(figsize=(10, 8))
sns.heatmap(data_correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#9. Plotting the relationship between price and Latest_Launch
#Converting the Latest_Launch data type to datetime
data["Latest_Launch"] = pd.to_datetime(data["Latest_Launch"])
```
